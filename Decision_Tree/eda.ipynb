```python
from scipy import stats
```

```python
contingency_table = pd.crosstab(df["credit_rating"], df["buys_computer"])
chi2, p, _, _ = stats.chi2_contingency(contingency_table)
print("Chi-square statistic:", chi2)
print("p-value:", p)
```
