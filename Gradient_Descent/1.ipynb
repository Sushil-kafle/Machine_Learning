fix the value of slope and change intercept, let's assume b=0

```python
y_pred = (m*x + b)
```

$$ slope = -2  {\sum_{i=1}^{n} y_i - m \cdot x_i -b   } $$


```python
b=0
m=78.35
loss_slope = -2 * np.sum(y-m*x.ravel() -b)
print(m ,b)
```

$$ b_{new} = b_{old} - \eta \cdot slope  $$
where:
- $ \eta $  is learning rate 
- slope is calculated slope at each iteration

let's aasume learning rate is 0.1


```python
lr =0.1
step_size = loss_slope*lr
b = b- step_size
step_size,b
```

```python
b=1000
lr =0.1
m=78.35
epoch =10

for i in range(epoch):
    loss_slope = -2 * np.sum(y-m*x.ravel() -b)
    b = b- loss_slope*lr
    y_pred = m*x+b
    plt.plot(x,y_pred)

plt.scatter(x,y)
```
